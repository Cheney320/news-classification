```python
#导入需要的库
import requests
from lxml import etree
from selenium import webdriver
import time
import warnings
warnings.filterwarnings("ignore")
# 构造要去爬取的url字典
url_dict = {'国际':'https://new.qq.com/ch/world/','娱乐':'https://new.qq.com/ch/ent/','体育':'https://new.qq.com/ch/sports/',\
            '军事':'https://new.qq.com/ch/milite/','科技':'https://new.qq.com/ch/tech/','财经':'https://new.qq.com/ch/finance/',\
            '汽车':'https://new.qq.com/ch/auto/','时尚':'https://new.qq.com/ch/fashion/','要闻':'https://new.qq.com/'}
url = input("爬取的新闻类别:")
driver = webdriver.PhantomJS()
driver.get(url_dict[url])
# 实现动态加载，爬取更多的url
for i in range(10):
    driver.find_element_by_id("load-more").click()
    time.sleep(2)
#lxml解析
selector = etree.HTML(driver.page_source)
# 提取新闻url
urls = selector.xpath('//h3/a/@href')
```

```python
titles = []  # 标题
articles = []  # 正文
times = []   # 时间
sources = []  # 来源（author）
comment_nums = []  # 评论数
comment_urls = []  # 评论
i = 1
print("{:*^50}".format("正在进入每个新闻url进行爬取"))
for url in urls:
    print("正在爬取第{}个页面：".format(i)+url)
    driver = webdriver.PhantomJS()
    driver.get(url)
    selector = etree.HTML(driver.page_source)
    title = selector.xpath('//div[@class="LEFT"]/h1/text()')
    if title == []:
        title = ''
    else:
        title = title[0]
    article = ''.join(selector.xpath('//p[@class="one-p"]/text()')).strip()
    year = selector.xpath('//div[@class="year through"]/span/text()')
    if year == []:
        year = ''
    else:
        year = year[0]
    md = ''.join(selector.xpath('//div[@class="md"]/text()'))
    hm = ''.join(selector.xpath('//div[@class="time"]/text()'))
    time = year + ' ' + md + ' ' + hm
    source = selector.xpath('//a[@class="author"]/div/text()')
    if source == []:
        source = ''
    else:
        source = source[0]
    comment_num = selector.xpath('//div[@class="text"]/i/text()')
    if comment_num == []:
        comment_num = ''
    else:
        comment_num = comment_num[0]
    comment_url = selector.xpath('//a[@class="comment-count"]/@href')
    if comment_url == []:
        comment_url = ''
    else:
        comment_url = comment_url[0]
    titles.append(title)
    articles.append(article)
    times.append(time)
    sources.append(source)
    comment_nums.append(comment_num)
    comment_urls.append(comment_url)
    i += 1
print("{:*^50}".format("爬取结束"))
```

```python
# 组合成数据帧
data_dic = {'title':titles, 'news_url':urls,'content':articles, 'date':times, 'source':sources, 'comment_num':comment_nums, \
            'comment_url':comment_urls, 'label':'要闻'}
import pandas as pd
data_df = pd.DataFrame(data_dic)
data_df
```

```python
#写入excel文件
data_df.to_excel('../../data/要闻5.xlsx',encoding='utf-8')
```
