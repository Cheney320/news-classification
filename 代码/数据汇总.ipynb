```python
import pandas as pd
import numpy as np
```

```python
# 将每个类别的新闻文件进行合并

tag_list = ['体育','军事','国际','娱乐','时尚','汽车','科技','财经']
all_data = {}
for tag in tag_list:
    data = []
    for i in range(1,6):
        df = pd.read_excel('../data/'+tag+str(i)+'.xlsx')
        data.append(df)
    result = pd.concat(data, ignore_index=True)
    all_data[tag] = result
```

```python
# 将每类数据drop缺失值，再合并

l = []
for k in all_data:
    result = all_data[k][['title','content','label']].dropna()
    result.index = range(len(result))
    l.append(result)
    
final_data = pd.concat(l, ignore_index=True)
```
