```python
import pandas as pd
import numpy as np
news = pd.read_excel('../../data/all_news.xlsx')
```

```python
news['content_length'] = news['content'].apply(lambda x:len(x.replace(' ','')))
```

```python
news.describe()
```

### 对长文本进行裁剪

```python
content = []
num = []
label = news['label'].tolist()
new_label = []
def split_content(x):
    content.extend(x.split('\n'))
    num.append(len(x.split('\n')))
_ = news[news['content_length'] > 100]['content'].apply(lambda x:split_content(x))
for i,j in zip(label, num):
    new_label.extend([i]*j)
```

```python
new_data = pd.DataFrame({'content':content, 'label':new_label})
```

```python
def clean_data(x):
    if x.isspace():
        x = np.NAN
    if len(str(x)) < 20:
        x = np.NAN
    return x
new_data['content'] = new_data['content'].apply(lambda x:clean_data(x))
```

```python
new_data.dropna(inplace=True)
```

### 随机打乱数据

```python
from sklearn.utils import shuffle
df_news = shuffle(new_data, random_state=0)
```

### 切分数据集

```python
X = df_news['content']
y = df_news['label']

# 切分为训练集和测试集
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

### 模型训练

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vect = TfidfVectorizer(max_features=95000, lowercase=False).fit(X_train)
X_train_tfidf = vect.transform(X_train)
X_test_tfidf = vect.transform(X_test)
```

```python
X_train_tfidf.shape
```

#### 朴素贝叶斯模型

```python
from sklearn.naive_bayes import MultinomialNB
clfrNB = MultinomialNB(alpha=0.02)
clfrNB.fit(X_train_tfidf, y_train)
```

```python
print(clfrNB.score(X_train_tfidf, y_train))
print(clfrNB.score(X_test_tfidf, y_test))
```

#### 逻辑回归模型

```python
from sklearn.linear_model import LogisticRegression
logis = LogisticRegression(C=5)
logis.fit(X_train_tfidf, y_train)
```

```python
print(logis.score(X_train_tfidf, y_train))
print(logis.score(X_test_tfidf, y_test))
```

#### SVM模型

```python
from sklearn.svm import SVC
svm = SVC(C=20000, probability=True)
svm.fit(X_train_tfidf, y_train)
```

```python
print(svm.score(X_train_tfidf, y_train))
print(svm.score(X_test_tfidf, y_test))
```

### 模型融合

```python
from sklearn.ensemble import VotingClassifier
eclf = VotingClassifier(estimators=[('nb', clfrNB), ('lg', logis)], voting='soft', weights=[1,2])
eclf.fit(X_train_tfidf, y_train)
print(eclf.score(X_train_tfidf, y_train))
print(eclf.score(X_test_tfidf, y_test))
```

```python
eclf2 = VotingClassifier(estimators=[('nb', clfrNB), ('lg', logis), ('svm', svm)], voting='soft', weights=[2,3,1])
eclf2.fit(X_train_tfidf, y_train)
```

```python
eclf2.predict_proba(X_test_tfidf)
```

```python
print(eclf2.score(X_train_tfidf, y_train))
print(eclf2.score(X_test_tfidf, y_test))
```

```python
from sklearn.metrics import classification_report

eclf2_predict = eclf2.predict(X_test_tfidf)
print(classification_report(y_test, eclf2_predict))
```

```python
eclf_predict = eclf.predict(X_test_tfidf)
print(classification_report(y_test, eclf_predict))
```
